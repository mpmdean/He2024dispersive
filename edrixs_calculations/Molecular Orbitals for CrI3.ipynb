Define the orbital IDs, the positions of each orbital, and the onsite energies for each orbital The cluster defined here takes I3 as the +x direction, I5 as the +y direction, I1 as the +z direction, I2 as the -x, I4 as the -y, and I6 as the -z. 


The index convention is as follows: 
1) "orb_dx2y2" is the $3d_{x^2-y^2}$ orbital on the Cr site, "orb_dz2r2" is the $3d_{3z^2-r^2}$ orbital on the Cr site, etc., 
2) "orb_px_x" is the $5p_x$ orbital on the ligand in the $+x$ direction (i.e. the I3), 
3) "orb_py_z" is the $5p_z$ orbital on the ligand in the $+y$ direction (i.e. the I5),
4) "orb_mx_y" is the $5p_y$ orbital on the ligand in the $-x$ direction (i.e. the I2), 
etc. 

i.e.,
 - px -> I3
 - mx -> I2
 - py -> I5
 - my -> I4
 - pz -> I1
 - mz -> I6

We will not actually need the positions for each orbital for this calculation but I have included this information in case we need it to check things later.

We now define the following ligand molecular orbitals
\begin{align}
L^\dagger_{x^2-y^2}&= \frac{1}{2}\left[-p^\dagger_{+x,x}+p^\dagger_{+y,y}+p^\dagger_{-x,x}-p^\dagger_{-y,y}\right], \\
L^\dagger_{3z^2-r^2}&= \frac{1}{2\sqrt{3}}\left[p^\dagger_{+x,x}+p^\dagger_{+y,y}-2p^\dagger_{+z,z}-p^\dagger_{-x,x}-p^\dagger_{-y,y}+2p^\dagger_{-z,z}\right], \\
L^\dagger_{xy}&= \frac{1}{2}\left[p^{\dagger}_{+x,y}+p^{\dagger}_{+y,x}-p^{\dagger}_{-x,y}-p^{\dagger}_{-y,x}\right], \\
L^\dagger_{xz}&= \frac{1}{2}\left[p^{\dagger}_{+x,z}+p^{\dagger}_{+z,x}-p^{\dagger}_{-x,z}-p^{\dagger}_{-z,x}\right],~\text{and} \\
L^\dagger_{yz}&= \frac{1}{2}\left[p^{\dagger}_{+y,z}+p^{\dagger}_{+z,y}-p^{\dagger}_{-y,z}-p^{\dagger}_{-z,y}\right]. \\
\end{align}
These are the standard combinations that hybridize with the central $d$ orbitals when the system has perfect cubic symmetry. 

If we define the row vector $\Psi^\dagger = \left[d^\dagger_{x^2-y^2}, d^\dagger_{3z^2-r^2}, d^\dagger_{xy}, d^\dagger_{xz}, d^\dagger_{yz}, L^\dagger_{x^2-y^2},L^\dagger_{3z^2-r^2},L^\dagger_{xy},L^\dagger_{xz},L^\dagger_{yz}\right]$ then the single-particle Hamiltonian is 
\begin{equation}
\mathcal{H} = \Psi^\dagger H \Psi
\end{equation}
where the matrix $H$ is given by 
\begin{equation}
